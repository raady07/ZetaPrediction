```python
# all imports 
import csv
import numpy as np
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix
import random
```

```python
# loading training data
file_path = '/home/dileep/Downloads/Home Challenge/data/2021-01-21_zeta-disease_training-data_dsi-take-home-challenge.csv'

data = [row for row in csv.reader(open(file_path))]
# column names
columns = data[0]
# labelled data
data = data[1:]
```

```python
# checking number of samples for each class
count_1 = 0
count_0 = 0 
for row in data:
    if row[-1] == '1':
        count_1 += 1
    else:
        count_0 += 1
        
print("0th class:", count_0)
print("1th class:", count_1)
```

```python
# selecting features - age[0], weight[1], blood_pressure[3], insulin[4], liver[5], cardio[6], smoking[7]
selected_data = [row[0: 2] + row[3:] for row in data]
```

```python
# making test and train data
random.shuffle(selected_data)

features = [row[:-1] for row in selected_data]
labels = [row[-1] for row in selected_data]

# 700 trainig samples 100 testing samples
features_training = features[0:701]
labels_training = labels[0:701]
features_testing = features[701:]
labels_testing = labels[701:]
```

```python
# training the SVM Model
model_linear = SVC(kernel='linear')
model_linear.fit(features_training, labels_training)

labels_predicted_linear = model_linear.predict(features_testing)

print("Accuracy: ", accuracy_score(labels_testing, labels_predicted_linear)*100)
print("Confusion Matrix: ", confusion_matrix(labels_testing, labels_predicted_linear))
```
